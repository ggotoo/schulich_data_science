# Understanding Data Distributions
### When are data distributions used?
- Data distributions are used extensively across different phases of data analysis and predictive modeling. They help in understanding the nature and structure of the data, which is crucial for:

- Descriptive Statistics: To summarize and describe data features.
- Inferential Statistics: To make predictions about a larger population based on sample data.
- Hypothesis Testing: To test assumptions and make decisions using data.
- Model Selection and Validation: To choose appropriate models and validate model assumptions.

### Why do we want to know the distribution of data?
Knowing the distribution of data is fundamental in statistics for several reasons:

- Model Fitting: Different statistical models and methods assume that data follows a specific distribution (e.g., linear regression assumes normally distributed errors). Knowing the data distribution helps in selecting the right model that fits the data appropriately.
- Decision Making: Understanding the distribution aids in making informed decisions based on the likelihood of various outcomes.
- Risk Assessment: In fields like finance and insurance, understanding distributions is key to assessing risks and tail events.
- Efficiency and Accuracy: Using the correct distribution improves the efficiency and accuracy of statistical estimates and predictions.


### How do we select which type of distribution to model data with?
Selecting the right distribution involves several steps and considerations:

1. Nature of Data:
- Continuous vs. Discrete: Continuous data might fit distributions like normal, exponential, or beta, while discrete data might fit Poisson or binomial distributions.
2. Data Characteristics:
- Symmetry and Skewness: Symmetric data may suggest a normal distribution, while skewed data might suggest exponential or gamma distributions.
- Boundedness: Data that are naturally bounded (e.g., percentages between 0 and 100) might fit a beta distribution.
3. Empirical Evidence: Plotting the data using histograms, box plots, or Q-Q plots helps visually assess the fit against known distribution shapes.
4. Statistical Tests: Tests like the Kolmogorov-Smirnov test, Shapiro-Wilk test, or Chi-Square goodness-of-fit test can statistically evaluate how well a distribution fits the data.
5. Business Understanding: The choice of distribution should also make sense logically based on the domain or business context.

### Distribution bounds

1. Unbounded Distributions

- Normal Distribution: The most famous example of an unbounded distribution, where data can theoretically range from negative to positive infinity, though practically most values lie within a few standard deviations of the mean.
- Poisson Distribution: Used for count data (non-negative integers), this distribution is unbounded above, meaning there is no upper limit to the values it can take, although the data must be greater than or equal to zero.

2. Semi-Bounded Distributions

- Exponential and Gamma Distributions: These are used for modeling non-negative continuous data (e.g., time until an event occurs). They are bounded below by zero but have no upper limit.
- Log-normal Distribution: This distribution is also bounded below by zero (since it's based on the exponential of a normal distribution) and extends infinitely upwards, suitable for data where values are positively skewed.

3.Fully Bounded Distributions

- Beta Distribution: Particularly useful because it is defined on the interval 
[0,1], making it ideal for modeling probabilities or proportions that are naturally restricted to this range.
- Uniform Distribution: Can be defined over any interval [ùëé,ùëè], where ùëé and ùëè are the minimum and maximum values, respectively. It assumes that all intervals of the same length are equally probable within the bounds.
- Dirichlet Distribution: A multivariate generalization of the beta distribution, used to model proportions that sum to one, and thus, each component is bounded between 0 and 1.

### How to select a distribution
1. Analyze the Data: Use plots and descriptive statistics to determine the range and distribution of the data.
2. Match Characteristics: Choose a distribution that accommodates the natural bounds of the data. For example, use a beta distribution for data that are percentages or proportions confined between 0 and 100%.
3. Model Testing: After selecting a distribution based on theoretical bounds, fit the model to the data and perform diagnostic tests to see if the chosen distribution provides a good fit. This can involve visual methods like Q-Q plots or statistical tests like the Anderson-Darling test.

### Normal Distribution
- Business Scenario: Estimating the average spending of customers at a retail store, where the spending patterns are symmetrically distributed around the mean.

- Data Analysis:

- Visualize Data: Plot a histogram and a Q-Q plot to check for the bell-shaped curve and normality.
- Calculate Descriptive Statistics: Compute mean, variance, skewness, and kurtosis to assess symmetry and peak.

#### Skewness:

- Definition: Skewness measures the asymmetry of the probability distribution of a real-valued random variable about its mean. The skewness can be:
- Positive (+): When the tail on the right side of the distribution is longer or fatter than the left side, indicating data are skewed right.
- Negative (‚àí): When the tail on the left side of the distribution is longer or fatter than the right side, indicating data are skewed left.
- Zero: When the tails on both sides of the mean balance out overall (which is true for a perfectly symmetric distribution like the normal distribution).

#### Kurtosis:

- Definition: Kurtosis measures the "tailedness" of the distribution. It is:
- Positive (+): More values in the tails than expected from the normal distribution (leptokurtic).
- Negative (‚àí): Fewer values in the tails than expected from the normal distribution (platykurtic).
- Zero: Similar to the normal distribution (mesokurtic), under Fisher‚Äôs definition where the normal distribution kurtosis is adjusted to zero.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate synthetic data
data = np.random.normal(loc=50, scale=10, size=1000)

# Histogram
plt.hist(data, bins=30, alpha=0.7, color='blue')
plt.show()

# Q-Q plot
stats.probplot(data, dist="norm", plot=plt)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate synthetic normal distribution data
data = np.random.normal(loc=0, scale=1, size=1000)

# Calculate Skewness and Kurtosis
skewness = stats.skew(data)
kurtosis = stats.kurtosis(data, fisher=True)  # Fisher's definition (normal == 0)

print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurtosis}")

# Histogram
plt.hist(data, bins=30, alpha=0.7, color='blue')
plt.title('Histogram of Normal Distribution')
plt.show()

# Q-Q plot
stats.probplot(data, dist="norm", plot=plt)
plt.title('Q-Q Plot of Normal Distribution')
plt.show()
```

### Beta Distribution
- Business Scenario: Analyzing the proportion of time spent on a particular activity by employees in a corporate setting, where proportions are naturally bounded between 0 and 1.

- Data Analysis:

- Visualize Data: Plot a histogram to check that data are bounded between 0 and 1.
- Assess Skewness: Check if the data is skewed towards the boundaries, which is typical for beta distributions.


```python
from scipy.stats import beta
import matplotlib.pyplot as plt

# Generate synthetic data
data = beta.rvs(a=2, b=5, size=1000)

# Histogram
plt.hist(data, bins=30, alpha=0.7, color='purple')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Parameters for left-skewed beta distribution
a = 5  # alpha > beta
b = 2  # beta

# Generate beta distribution data
data = beta.rvs(a, b, size=1000)

# Plot histogram
plt.hist(data, bins=30, alpha=0.7, color='purple')
plt.title(f'Beta Distribution: alpha={a}, beta={b} (Left-Skewed)')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
data = beta.rvs(a=2, b=5, size=1000)  # For left-skewed
stats.probplot(data, dist="beta", sparams=(2,5), plot=plt)
plt.title('Q-Q Plot for Beta Distribution')
plt.show()
```

### Statistical Tests based on Distribution
1. Normal Distribution
- Parametric Tests: These tests assume that the data follows a normal distribution.

- t-Test: Used to compare the means between two groups. Applicable for independent or paired samples.
- ANOVA (Analysis of Variance): Used to compare the means of three or more groups. Extensions like MANOVA (Multivariate Analysis of Variance) are used for multiple dependent variables.
- Linear Regression: Used to model relationships between a dependent variable and one or more independent variables when the residuals are normally distributed.

2. Non-Normal Distributions
When the data do not follow a normal distribution, non-parametric tests, which do not assume an underlying distribution, are appropriate.

##### For Skewed Distributions (e.g., Exponential, Log-Normal)
- Mann-Whitney U Test: A non-parametric test for assessing whether two independent samples originate from the same distribution. Useful when the data are not normally distributed.
- Kruskal-Wallis H Test: Similar to ANOVA, but for non-parametric data. Used to compare the medians of three or more groups.
##### For Count Data (e.g., Poisson)
- Chi-Square Test of Independence: Used to determine if there is a significant association between two categorical variables.
- Fisher‚Äôs Exact Test: Used for small sample sizes, providing an exact p-value for the association between categorical variables.

3. Bounded Distributions (e.g., Beta, Uniform)
- Kolmogorov-Smirnov Test: A non-parametric test that compares the cumulative distributions of two data sets, or a sample with a known distribution.
- Anderson-Darling Test: More sensitive than the Kolmogorov-Smirnov test at the tails of the distribution. It is used to check if a sample comes from a specific distribution.

4. Binary or Nominal Distributions (e.g., Binomial)
- Logistic Regression: Used to model binary outcome variables (e.g., success/failure) based on one or more predictor variables.
- Chi-Square Goodness of Fit Test: Tests whether the observed proportions for a categorical variable differ from expected proportions.

#### How to Choose the Correct Test
1. Assess the Distribution: Use visual plots (histograms, Q-Q plots), descriptive statistics (skewness, kurtosis), and tests (Shapiro-Wilk test for normality) to understand the data distribution.
2. Consider the Data Type: Determine if the data are independent or paired, and if the variables are categorical, ordinal, or continuous.
Determine the Objective: Decide if you are comparing means/medians, assessing correlations, or testing associations.
3. Check Sample Size and Data Conditions: Some tests, like Fisher‚Äôs Exact Test, are more appropriate for small sample sizes, while others, like the t-Test, require larger samples to be effective.

1. t-Test (Normal Distribution)
Business Scenario: Comparing average monthly sales figures between two different stores of the same retail chain to determine if one store consistently outsells the other.



```python
import numpy as np
from scipy.stats import ttest_ind

# Generate synthetic sales data
store_a_sales = np.random.normal(30000, 5000, 100)  # mean = 30000, std = 5000
store_b_sales = np.random.normal(35000, 5000, 100)  # mean = 35000, std = 5000

# Perform a t-test
t_stat, p_value = ttest_ind(store_a_sales, store_b_sales)
print(f'T-statistic: {t_stat}, P-value: {p_value}')
```

2. Mann-Whitney U Test (Skewed Distribution)
Business Scenario: Determining whether customer satisfaction ratings between two different service teams differ, where ratings are ordinal and not normally distributed.

```python
from scipy.stats import mannwhitneyu

# Generate synthetic satisfaction ratings
team_a_ratings = np.random.exponential(scale=3, size=50)
team_b_ratings = np.random.exponential(scale=5, size=50)

# Mann-Whitney U Test
u_stat, p_value = mannwhitneyu(team_a_ratings, team_b_ratings)
print(f'U-statistic: {u_stat}, P-value: {p_value}')
```

3. Kruskal-Wallis H Test (Multiple Groups, Non-Normal)
Business Scenario: Testing if three different marketing strategies result in different levels of customer engagement, measured through a non-normally distributed engagement score.

```python
from scipy.stats import kruskal

# Generate synthetic engagement scores
strategy_1 = np.random.gamma(shape=2, scale=2, size=30)
strategy_2 = np.random.gamma(shape=2, scale=2, size=30)
strategy_3 = np.random.gamma(shape=2, scale=2, size=30)

# Kruskal-Wallis H Test
h_stat, p_value = kruskal(strategy_1, strategy_2, strategy_3)
print(f'H-statistic: {h_stat}, P-value: {p_value}')
```

4. Chi-Square Test of Independence (Categorical Data)
Business Scenario: Evaluating if there is a relationship between two categorical variables: customer gender and preference for a new product line.

```python
from scipy.stats import chi2_contingency

# Generate a contingency table
table = [[10, 20, 30], [20, 20, 20]]
chi2, p, dof, expected = chi2_contingency(table)
print(f'Chi2 statistic: {chi2}, P-value: {p}')
```
