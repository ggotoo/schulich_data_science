```python
df = df.drop(columns=['sanction_threat_trade', 'sanction_threat', 'imp_o_98', 'sanction_imposition_trade', 'sanction_imposition', 'hostility_level_o', 'hostility_level_d'])
```

```python
df.info()
```

```python
df.head()
```

```python
for column in df.columns:
    if df[column].dtype in ['float64', 'int64']:
        df[column].fillna(df[column].median(), inplace=True)
    else:
        df[column].fillna(df[column].mode()[0], inplace=True)
```
