```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
data = np.random.normal(loc=50, scale=10, size=1000)
```

```python
plt.hist(data, bins=20, alpha=0.7, color='blue')
```

```python
stats.probplot(data, dist='norm', plot=plt);
```

```python
print(stats.skew(data))
print(stats.kurtosis(data, fisher=True))
```

```python
#Poisson Distribution
from scipy.stats import poisson
```

```python
data = poisson.rvs(mu=10, size=1000)
plt.hist(data, bins=30, alpha=0.7, color='green');
```

```python
# exponential distribution
data = np.random.exponential(scale=30, size=1000)
plt.hist(data, bins=50, color='red');
```

```python
print(stats.skew(data))
print(stats.kurtosis(data, fisher=True))
```

```python
# t-test
store_a_sales = np.random.normal(30000, 5000, 100)
store_b_sales = np.random.normal(35000, 5000, 100)
```

```python
from scipy.stats import ttest_ind
t_stat, p_value = ttest_ind(store_a_sales, store_b_sales, equal_var=False, alternative='greater')
```

```python
from scipy.stats import ttest_ind
t_stat, p_value = ttest_ind(store_a_sales, store_b_sales, equal_var=True, alternative='two-sided')
```

```python
print(p_value)
```

```python
# mann whitney
from scipy.stats import mannwhitneyu
u_stat, p_value = mannwhitneyu(store_a_sales, store_b_sales)
print(p_value)
```

```python
df.groupby('Customer_ID')[['Customer_Age_Months', 'Gender', 'Campaign_Exposure','Transaction_Category', 'Purchase_Amount', 'Premium_Product']].agg({'Customer_Age_Months':'max', 'Campaign_Exposure':'max', 'Purchase_Amount':'mean', 'Premium_Product':'mean'})
cust_level = df.groupby('Customer_ID')[['Customer_Age_Months', 'Gender', 'Campaign_Exposure','Transaction_Category', 'Purchase_Amount', 'Premium_Product']].agg({'Customer_Age_Months':'max', 'Campaign_Exposure':'max', 'Purchase_Amount':'mean', 'Premium_Product':'mean'})
```

```python
campaign_cx = cust_level[cust_level['Campaign_Exposure']==1]
reg_cx = cust_level[cust_level['Campaign_Exposure']==0]
```

```python
sns.boxplot(x = 'Campaign_Exposure', y = 'Purchase_Amount', data=cust_level)
```

```python
sns.boxplot(x = 'Campaign_Exposure', y = 'Premium_Product', data=cust_level)
```

```python
# ttest first
t_stat, p_value = ttest_ind(campaign_cx['Purchase_Amount'], reg_cx['Purchase_Amount'])
print(p_value)
```

```python
# ttest first
t_stat, p_value = ttest_ind(campaign_cx['Purchase_Amount'], reg_cx['Purchase_Amount'], alternative='greater')
print(p_value)
```

```python
t_stat, p_value = ttest_ind(campaign_cx['Premium_Product'], reg_cx['Premium_Product'], alternative='greater')
print(p_value)
```
